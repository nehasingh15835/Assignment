```python
#Question 4: Create a bar plot to visualize the average selling price for each seller_typeand write one observation.

import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv(r"C:\Users\Neha Singh\Downloads\Bike Details.csv")


# Group by seller_type and calculate mean selling_price
avg_prices = df.groupby("seller_type")["selling_price"].mean()

# Plot bar chart
plt.figure(figsize=(7,5))
avg_prices.plot(kind="bar", color="skyblue", edgecolor="black")

plt.title("Average Selling Price by Seller Type", fontsize=14)
plt.xlabel("Seller Type", fontsize=12)
plt.ylabel("Average Selling Price (₹)", fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```

```python
#Question 6: Use the IQR method to detect and remove outliers from the km_drivencolumn. Show before-and-after summary statistics.

import pandas as pd

# Load dataset
df = pd.read_csv(r"C:\Users\Neha Singh\Downloads\Bike Details.csv")

# Summary statistics before removing outliers
print("Summary statistics BEFORE removing outliers:\n")
print(df["km_driven"].describe())

# Step 1: Calculate Q1, Q3, and IQR
Q1 = df["km_driven"].quantile(0.25)
Q3 = df["km_driven"].quantile(0.75)
IQR = Q3 - Q1

# Step 2: Define bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print("\nLower Bound:", lower_bound)
print("Upper Bound:", upper_bound)

# Step 3: Remove outliers
df_clean = df[(df["km_driven"] >= lower_bound) & (df["km_driven"] <= upper_bound)]

# Summary statistics after removing outliers
print("\nSummary statistics AFTER removing outliers:\n")
print(df_clean["km_driven"].describe())
```

```python
#Question 9: Generate a heatmap of the correlation matrix for all numeric columns.What correlations stand out the most?

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv(r"C:\Users\Neha Singh\Downloads\Bike Details.csv")


# Compute correlation matrix (numeric columns only)
corr_matrix = df.corr(numeric_only=True)

# Plot heatmap
plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)

plt.title("Correlation Heatmap of Numeric Columns", fontsize=14)
plt.show()
```

```python

#Question 10: Summarize your findings in a brief report:
# What are the most important factors affecting a bike's selling price?
# Mention any data cleaning or feature engineering you performed.

# 📝 Report on Bike Price Analysis

### 🔹 Key Factors Affecting a Bike’s Selling Price

1. Ex-Showroom Price

   * Strongest positive correlation with selling price.
   * Higher showroom value → higher resale price.

2. Year of Manufacture

   * Newer bikes retain higher value.
   * Older bikes show lower resale prices, but exceptions exist (e.g., premium/classic models).

3. Kilometers Driven (km\_driven)

   * Negative correlation with selling price.
   * Bikes with more usage generally sell for less.

4. Ownership Type

   * First-owner bikes usually have higher value than 2nd/3rd owner bikes.

5. Seller Type

   * Dealers tend to list bikes at higher prices compared to individuals.

---

# Data Cleaning and Feature Engineering Steps

1. Handling Missing Values

   * Replaced missing numerical values (e.g., `ex_showroom_price`) with the median.
   * Replaced missing categorical values with the mode.
   * In some cases, missing values were replaced with `"NA"` to retain information.

2. Outlier Treatment

   * Used the IQR method to detect and remove unrealistic outliers in `km_driven`.
   * Reduced extreme values to make statistics more reliable.

3. Encoding Categorical Variables

   * Converted `seller_type` into numeric format using One-Hot Encoding for machine learning.

4. Exploratory Data Analysis (EDA)

   * Plotted histograms to study price distribution (skewed towards lower-priced bikes).
   * Created bar plots to compare average prices by seller type and ownership.
   * Scatter plots showed the downward trend of price with age.
   * Correlation heatmap highlighted key relationships.

# Overall Findings

* Ex-showroom price and year are the strongest predictors of resale value.
* Usage (km\_driven) and ownership history also significantly influence price.
* Data cleaning (handling missing values, removing outliers) improved data quality.
* Feature engineering (one-hot encoding) made the dataset ready for machine learning models.


```
