```python

#Question1:Explain the differences between AI, ML, Deep Learning (DL), and Data Science (DS).
1. Artificial Intelligence (AI)
Definition: The broadest concept â€“ AI is about creating machines or systems that can perform tasks that normally require human intelligence (e.g., reasoning, decision-making, problem-solving, understanding language).
Goal: Make computers "think" and act intelligently.
Examples:
Chatbots like Siri or Alexa
Self-driving cars
Fraud detection systems

2. Machine Learning (ML)
Definition: A subset of AI where machines learn patterns from data instead of being explicitly programmed with rules.
Goal: Improve performance automatically through experience (data).
Techniques: Supervised, Unsupervised, Reinforcement learning.
Examples:
Email spam filters
Netflix movie recommendations
Predicting stock prices

3. Deep Learning (DL)

Definition: A subset of ML that uses Artificial Neural Networks (inspired by the human brain) with many layers ("deep" networks).
Goal: Handle large, complex datasets (especially images, text, audio) more effectively than traditional ML.
Examples:
Facial recognition on Facebook/Instagram
Voice assistants like Google Assistant
Medical image diagnosis

3. Data Science (DS)
Definition: A multidisciplinary field that uses statistics, data analysis, visualization, and ML to extract insights and knowledge from structured & unstructured data.
Goal: Turn raw data into actionable insights for decision-making.
Key Components: Data collection, cleaning, analysis, visualization, ML models.
Examples:
Analyzing customer behavior for marketing
Business dashboards & reports
Predicting sales trends

```

```python
#Q2: What are the types of machine learning? Describe each with one real-world example.

1. Supervised Learning
Definition: The model is trained on a labeled dataset (input + correct output). The algorithm learns the mapping from input â†’ output.
Goal: Predict outcomes for new, unseen inputs.
Techniques:
Classification (predict categories)
Regression (predict continuous values)

Real-World Example:
Email Spam Detection â†’ The algorithm is trained with emails labeled as â€œspamâ€ or â€œnot spamâ€. When a new email arrives, it predicts whether it is spam.

2. Unsupervised Learning
Definition: The model is trained on unlabeled data (only input, no output). It tries to find hidden patterns, structures, or groupings in the data.
Goal: Discover insights or groupings without predefined labels.
Techniques:
Clustering (grouping data)
Dimensionality Reduction (simplifying data while keeping meaning)

Real-World Example:
Customer Segmentation in Marketing â†’ An e-commerce company uses unsupervised ML to group customers by behavior (e.g., frequent buyers, discount seekers
one-time buyers) for targeted advertising.

3. Reinforcement Learning (RL)

Definition: The model learns by interacting with an environment. It takes actions and receives rewards (positive) or penalties (negative).
Goal: Learn the best sequence of actions to maximize long-term rewards.
Techniques: Q-Learning, Deep Reinforcement Learning.

Real-World Example:

Self-Driving Cars â†’ The carâ€™s AI learns how to drive safely by receiving rewards for staying in the lane, avoiding crashes, and penalties for violations.

4. Semi-Supervised Learning (Hybrid approach â€“ optional but important)
Definition: Uses a small amount of labeled data + a large amount of unlabeled data. The model learns better than using just one type of data.
Goal: Improve learning accuracy when labeling data is expensive or time-consuming.

Real-World Example:

Medical Diagnosis â†’ Doctors provide a few labeled X-ray scans (diseased/healthy), and the algorithm learns from them while using thousands of
unlabeled scans to improve accuracy.
```

```python
#Q3. Define overfitting, underfitting, and the bias-variance tradeoff in machine learning?
1. Overfitting
Definition: When a model learns too much from training data, including noise and random fluctuations, making it perform well on training data 
but poorly on new/unseen data.
Cause: Model is too complex (too many parameters, deep trees, too many layers).
Example: A student memorizes all questions/answers from a practice exam but fails when faced with new questions in the real exam.
Symptom:
Very low training error 
Very high test/validation error 

2.Underfitting
Definition: When a model is too simple to capture the patterns in the data, resulting in poor performance on both training and test data.
Cause: Model is too weak (not enough parameters, shallow trees, linear model on non-linear data).
Example: A student only skims the textbook without practicing; they fail both practice and real exams.
Symptom:
High training error 
High test error 

3. Bias-Variance Tradeoff
Definition: The balance between two types of errors that affect model performance:
Bias â†’ Error from overly simplifying the model (underfitting).
Model makes strong assumptions, missing real patterns.
Variance â†’ Error from too much complexity (overfitting).
Model is overly sensitive to small fluctuations in the training set.
Goal: Find the sweet spot between bias and variance â†’ good generalization.
```

```python
#Q4 What are outliers in a dataset, and list three common techniques for handling them.
What are Outliers?
Definition: Outliers are data points that deviate significantly from the majority of the dataset.
They can be much higher or lower than other observations.
Cause: Measurement errors, data entry mistakes, rare events, or genuine extreme values.
Impact:
Can skew mean & standard deviation
Mislead machine learning models
Affect statistical tests
Example: In a dataset of peopleâ€™s ages [22, 24, 25, 23, 120], the value 120 is likely an outlier.

Three Common Techniques to Handle Outliers
1. Removal (Deleting Outliers)
When to use: If outliers are due to errors or irrelevant anomalies.
Method: Remove values that lie beyond a certain threshold (e.g., Â±3 standard deviations from mean).
Example: Removing "age = 120" if dataset is about college students.

2. Transformation (Reducing Impact)
When to use: If outliers are genuine but too extreme, apply mathematical transformations.
Techniques:
Log Transformation
Square Root Transformation
Normalization/Standardization
Example: Converting salary data [30K, 35K, 40K, 5M] using log transformation reduces the gap between 40K and 5M.
Capping / Winsorization
When to use: If you want to keep outliers but limit their influence.
Method: Replace extreme values with a boundary (e.g., 1st and 99th percentile).
Example: If income > 1,000,000, cap it to 1,000,000.
    
```

```python
#Q5 Explain the process of handling missing values and mention oneimputation technique for numerical and one for categorical data.
Process of Handling Missing Values

1.Identify Missing Data

Check for NaN, null, or blank values.

In Python â†’ df.isnull().sum()

2.Understand the Cause

Data entry errors, sensor failures, skipped survey questions, etc.

3.Decide on a Strategy

Remove: Drop rows/columns with too many missing values (if >50%).

Impute (Fill): Replace missing values with a suitable estimate.

Predict: Use ML models to estimate missing values.

Imputation Techniques
ðŸ”¹ For Numerical Data

Mean / Median Imputation
Replace missing values with the mean (for normal distributions) or median (for skewed data).

Example:
Dataset: [5, 8, NaN, 10, 7]

Imputed with mean = 7.5 â†’ [5, 8, 7.5, 10, 7]
For Categorical Data

Mode Imputation
Replace missing values with the most frequent category.

Example:
Dataset: ['Red', 'Blue', NaN, 'Red', 'Green']
Mode = 'Red' â†’ ['Red', 'Blue', 'Red', 'Red', 'Green']
```

```python
#Q8 Write a Python script to:
#Generate 1000 samples from a normal distribution.
#Introduce 50 random missing values.
#Fill missing values with the column mean.
#Plot a histogram before and after imputation

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Generate 1000 samples from a normal distribution
np.random.seed(42)  # for reproducibility
data = np.random.normal(loc=50, scale=10, size=1000)  # mean=50, std=10

# Step 2: Introduce 50 random missing values
data_missing = data.copy()
missing_indices = np.random.choice(1000, 50, replace=False)  # choose 50 random indices
data_missing[missing_indices] = np.nan

# Step 3: Fill missing values with column mean
data_filled = pd.Series(data_missing).fillna(np.nanmean(data_missing))

# Step 4: Plot histograms before and after imputation
plt.figure(figsize=(12, 5))

# Before imputation
plt.subplot(1, 2, 1)
plt.hist(data_missing[~np.isnan(data_missing)], bins=30, edgecolor='black')
plt.title("Before Imputation (with Missing Values)")
plt.xlabel("Value")
plt.ylabel("Frequency")

# After imputation
plt.subplot(1, 2, 2)
plt.hist(data_filled, bins=30, edgecolor='black')
plt.title("After Imputation (Mean Filled)")
plt.xlabel("Value")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```

```python
#Q10 You are working as a data scientist for a retail company. You receive a customertransaction dataset that contains:
#Missing ages,
#Outliers in transaction amount,
#A highly imbalanced target (fraud vs. non-fraud),
#Categorical variables like payment method.
#Explain the step-by-step data preparation plan youâ€™d follow before training a machine learning
#model. Include how youâ€™d address missing data, outliers, imbalance, and encoding


# Data-prep + resampling + training pipeline
# Works whether or not imbalanced-learn (imblearn) is installed.
# Assumes `df` is your DataFrame and contains:
# ['transaction_amount', 'age', 'tenure_days', 'payment_method', 'country', 'is_fraud']

import numpy as np
import pandas as pd
from collections import Counter

from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, RobustScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, average_precision_score, precision_recall_curve
from sklearn.utils import resample

# Try to import imblearn (SMOTE); otherwise we'll fallback to random oversampling.
try:
    from imblearn.over_sampling import SMOTE
    IMBLEARN_AVAILABLE = True
except Exception:
    IMBLEARN_AVAILABLE = False

# ---------------------------
# 1) Helper functions
# ---------------------------
def print_dataset_overview(df, target_col='is_fraud'):
    print("=== Dataset overview ===")
    print("Shape:", df.shape)
    print("\nMissing values per column:")
    print(df.isnull().sum())
    print("\nTarget distribution:")
    print(df[target_col].value_counts(normalize=False))
    print(df[target_col].value_counts(normalize=True))
    print("========================\n")

def cap_outliers_iqr(train_series, clip_series, multiplier=1.5):
    """
    Compute IQR-based bounds on train_series and clip clip_series to those bounds.
    Returns clipped version of clip_series and the bounds used.
    """
    q1 = train_series.quantile(0.25)
    q3 = train_series.quantile(0.75)
    iqr = q3 - q1
    lower = q1 - multiplier * iqr
    upper = q3 + multiplier * iqr
    clipped = clip_series.clip(lower=lower, upper=upper)
    return clipped, (lower, upper)

def random_oversample_numpy(X, y, minority_label=1, random_state=42):
    """
    X: numpy array (n_samples, n_features)
    y: 1D array-like labels
    Returns oversampled X_res, y_res where minority class is upsampled to match majority.
    """
    y = np.asarray(y)
    X = np.asarray(X)
    classes, counts = np.unique(y, return_counts=True)
    if len(classes) != 2:
        raise ValueError("This simple random_oversample_numpy supports binary only.")
    majority_label = classes[np.argmax(counts)]
    minority_label_found = classes[np.argmin(counts)]
    if minority_label_found != minority_label:
        minority_label = minority_label_found

    X_min = X[y == minority_label]
    X_maj = X[y == majority_label]
    y_min = y[y == minority_label]
    y_maj = y[y == majority_label]

    # Upsample minority to match majority
    X_min_upsampled = resample(X_min, replace=True, n_samples=len(X_maj), random_state=random_state)
    y_min_upsampled = np.array([minority_label] * len(X_maj))

    X_res = np.vstack([X_maj, X_min_upsampled])
    y_res = np.concatenate([y_maj, y_min_upsampled])

    # Shuffle
    idx = np.arange(len(y_res))
    np.random.seed(random_state)
    np.random.shuffle(idx)
    return X_res[idx], y_res[idx]

# ---------------------------
# 2) Main function
# ---------------------------
def prepare_and_train(df,
                      target_col='is_fraud',
                      num_cols=['transaction_amount', 'age', 'tenure_days'],
                      cat_cols=['payment_method', 'country'],
                      test_size=0.2,
                      random_state=42):
    """
    df: DataFrame containing features + target
    returns: trained model, preprocessor, and evaluation on test set
    """

    # 2.0 Basic checks / overview
    print_dataset_overview(df, target_col=target_col)

    # 2.1 Train/Test split (stratified)
    X = df[num_cols + cat_cols].copy()
    y = df[target_col].copy()
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=test_size, stratify=y, random_state=random_state
    )
    print("After split â€” train shape:", X_train.shape, "test shape:", X_test.shape)
    print("Train target distribution:", Counter(y_train))
    print("Test target distribution:", Counter(y_test))
    print()

    # 2.2 Outlier handling for `transaction_amount` using IQR on TRAIN only
    print("Computing outlier caps for 'transaction_amount' on TRAIN set using IQR...")
    capped_train_amt, (lower, upper) = cap_outliers_iqr(X_train['transaction_amount'], X_train['transaction_amount'])
    X_train = X_train.copy()
    X_train['transaction_amount'] = capped_train_amt
    # Apply same caps to test set (clip using train bounds)
    X_test = X_test.copy()
    X_test['transaction_amount'] = X_test['transaction_amount'].clip(lower=lower, upper=upper)
    print(f"Applied caps: lower={lower:.2f}, upper={upper:.2f}\n")

    # 2.3 Preprocessing pipelines
    numeric_pipeline = Pipeline(steps=[
        ('imputer', SimpleImputer(strategy='median')),   # median handles skew & is robust to outliers
        ('scaler', RobustScaler())                       # robust scaling reduces influence of extreme values
    ])

    categorical_pipeline = Pipeline(steps=[
        ('imputer', SimpleImputer(strategy='most_frequent')),  # fill missing categories with mode
        ('ohe', OneHotEncoder(handle_unknown='ignore', sparse=False))  # low-cardinality OHE
    ])

    preprocessor = ColumnTransformer(transformers=[
        ('num', numeric_pipeline, num_cols),
        ('cat', categorical_pipeline, cat_cols)
    ], remainder='drop')

    # Fit preprocessor on TRAIN data and transform train/test
    print("Fitting preprocessor on TRAIN and transforming data...")
    X_train_trans = preprocessor.fit_transform(X_train)
    X_test_trans = preprocessor.transform(X_test)
    print("Transformed shapes:", X_train_trans.shape, X_test_trans.shape)
    print()

    # Check result distribution BEFORE resampling
    print("Before resampling (train):", Counter(y_train))

    # 2.4 Resampling to handle class imbalance
    if IMBLEARN_AVAILABLE:
        print("imblearn detected: Using SMOTE to oversample minority class (on transformed numeric matrix).")
        smote = SMOTE(sampling_strategy=0.5, random_state=random_state) 
        # sampling_strategy=0.5: minority will become ~50% of majority (you can adjust)
        X_res, y_res = smote.fit_resample(X_train_trans, y_train)
        print("After SMOTE resample:", Counter(y_res))
    else:
        print("imblearn not available: falling back to simple random oversampling of minority to match majority.")
        X_res, y_res = random_oversample_numpy(X_train_trans, y_train.values, random_state=random_state)
        print("After random oversample:", Counter(y_res))

    # 2.5 Fit a classifier (cost-sensitive logistic regression)
    clf = LogisticRegression(max_iter=2000, class_weight='balanced', random_state=random_state)
    print("Training classifier on resampled data...")
    clf.fit(X_res, y_res)

    # 2.6 Evaluate on TEST set (do not resample test)
    print("\n=== Evaluation on TEST set ===")
    y_pred = clf.predict(X_test_trans)
    y_prob = clf.predict_proba(X_test_trans)[:, 1] if hasattr(clf, "predict_proba") else None

    print("Classification report:")
    print(classification_report(y_test, y_pred, digits=4))

    if y_prob is not None:
        ap = average_precision_score(y_test, y_prob)
        print(f"Average precision (PR-AUC): {ap:.4f}")

    print("Confusion matrix:")
    print(confusion_matrix(y_test, y_pred))

    return {
        'preprocessor': preprocessor,
        'model': clf,
        'X_test_trans': X_test_trans,
        'y_test': y_test,
        'y_pred': y_pred,
        'y_prob': y_prob
    }

# ---------------------------
# 3) Example usage
# ---------------------------
if __name__ == "__main__":
    # Example: if you already have `df`, call prepare_and_train(df)
    # For demonstration, I'll create a tiny dummy df with the columns (in practice use your real data)
    # Remove or replace this block if you are calling the function with your real df.

    # ---- demo synthetic small example (remove for real usage) ----
    np.random.seed(42)
    n = 1000
    df_demo = pd.DataFrame({
        'transaction_amount': np.concatenate([np.random.gamma(2, 50, size=n-20), np.random.normal(5000, 500, size=20)]),
        'age': np.random.choice([np.nan] * 50 + list(range(18, 70)), size=n),  # include some missing ages
        'tenure_days': np.random.randint(1, 2000, size=n),
        'payment_method': np.random.choice(['card', 'netbanking', 'wallet'], size=n),
        'country': np.random.choice(['IN', 'US', 'UK', 'CA'], size=n),
        'is_fraud': np.concatenate([np.zeros(n-30), np.ones(30)])  # imbalanced: small fraud proportion
    })

    results = prepare_and_train(df_demo, target_col='is_fraud')
    # You can inspect results['model'], results['preprocessor'], etc.
```
