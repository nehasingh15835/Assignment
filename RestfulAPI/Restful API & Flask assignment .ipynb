```python
# Restful API & Flask Assignment 

#Q1 What is a RESTful API?
API:
API stands for Application Programming Interface.
It’s like a menu in a restaurant: it tells you what you can order (do), and how to ask for it.
Example: When you use a weather app, the app asks a weather API to get the current temperature.
RESTful:
REST stands for REpresentational State Transfer.
It’s a set of rules that makes APIs simple, fast, and organized.
A RESTful API follows these rules to help systems communicate easily using the internet.

A RESTful API lets programs talk to each other over the web.
It uses HTTP methods like GET, POST, PUT, DELETE.
It helps build simple and organized web services.
It’s like a messenger between a user and a system.


#Q2 Explain the concept of API specification?
An API specification is like a detailed blueprint or instruction manual that explains how to use an API.

 API specification is a detailed guide that explains how an API works.
It shows what you can do, what data to send, and what you’ll get back.
It helps ensure everything is clear and consistent for anyone using the API.

#Q3 What is Flask, and why is it popular for building APIs?
Flask is a lightweight web framework written in Python that helps you build web applications and APIs quickly and easily.
Think of Flask as a toolkit that gives you everything you need to build a simple website or a backend service, especially for RESTful APIs.
Why popular:
Lightweight	
Flexible	
REST support	
Extensions	
Community

#Q4 What is routing in Flask?
Routing in Flask is the process of connecting a URL (web address) to a specific function in your Python code.
This is how Flask knows what to do when someone visits a certain webpage or API endpoint.
Route     	       A URL path handled by Flask
@app.route  	   Used to define which function responds to which path
Static route	   Fixed URL like /about
Dynamic route	   URL with variable like /user/<name>
Methods	Controls   allowed actions like GET or POST

#Q5 How do you create a simple Flask application?
1.  Install Flask
2.	Create a Python file
3.	Set up routes using @app.route
4.	Run the app
5.	Open it in your browser
    
#Q6 What are HTTP methods used in RESTful APIs?
In RESTful APIs, HTTP methods define what kind of action you want to perform on data (like reading, creating, updating, or deleting).
 HTTP methods are like commands you give to a REST API:
1."GET this info",
2."POST a new item",
3."PUT updated info",
4."PATCH just one field",
5.or "DELETE it".
 
#Q7 What is the purpose of the @app.route() decorator in Flask?
The @app.route() decorator in Flask is used to define a route — that is, it tells Flask which URL should trigger which function.
 Define URL paths
 Handle different HTTP methods (GET, POST, etc.)
Create dynamic routes (with variables in the URL)

#Q8 What is the difference between GET and POST HTTP methods?
GET and POST are two of the most commonly used HTTP methods in web development and RESTful APIs. They both send requests to a server, 
but they behave differently.
GET – "Give me data!"
used to: Retrieve data from the server.
Does not change anything on the server.
Data is sent in the URL (query string).
Less secure (data visible in browser history).
Can be bookmarked and cached.

POST – "Send this data!"
Used to: Send or submit data to the server.
Often used to create or update something.
Data is sent in the request body (not in the URL).
More secure than GET (not shown in browser history).
Cannot be bookmarked.
    
#Q9 How do you handle errors in Flask APIs?
When building Flask APIs, it's important to handle errors gracefully so that users or clients (like mobile apps) get clear and helpful 
error messages instead of crashes or confusing responses.
    
Flask lets you handle errors using:
1.try-except for logic errors
2.@app.errorhandler for HTTP errors
3.abort() for custom errors
    
#Q10 How do you connect Flask to a SQL database?
To connect Flask to a SQL database (like SQLite, MySQL, or PostgreSQL), you typically use a tool called SQLAlchemy, which makes working with databases 
in Python much easier and cleaner.

#Q11 What is the role of Flask-SQLAlchemy?
Flask-SQLAlchemy is an extension for Flask that makes it easy to work with SQL databases (like SQLite, MySQL, PostgreSQL)
using Python classes and objects.
Easy to connect to a database
Easy to define tables using Python
Easy to run queries without writing SQL

#Q12 What are Flask blueprints, and how are they useful?
Flask Blueprints are a way to organize your Flask app into reusable, modular pieces — like breaking your app into smaller sections.
They’re especially useful when your app grows and you don’t want all your code in one huge file.
Blueprint()	              Creates a modular group of routes and logic
@blueprint.route()	      Defines routes in the blueprint
app.register_blueprint()  Adds the blueprint to the main app
url_prefix                Adds a prefix to all routes in the blueprint

#Q13 What is the purpose of Flask's request object?
The request object in Flask is used to access data sent by the client to the server.
Whenever someone sends a request to your Flask app (like submitting a form, sending JSON data, or clicking a link), the request object 
lets you read all that incoming data.
    
#Q14 How do you create a RESTful API endpoint using Flask?
A RESTful API endpoint in Flask is simply a route that responds to HTTP methods like GET, POST, PUT, or DELETE, and returns data — 
usually in JSON form.
    
#Q15 What is the purpose of Flask's jsonify() function?
The jsonify() function in Flask is used to convert Python data (like dictionaries or lists) into JSON format and return it as a proper HTTP response.
Purpose:
Converts Python data (dict, list, etc.) into valid JSON
Automatically sets the Content-Type header to application/json
Handles Unicode and special characters properly
Supports status codes and headers

#Q16 Explain Flask’s url_for() function?
Flask’s url_for() function is used to build URLs dynamically by referring to the name of a function (route handler) — not by hardcoding the URL path.
Avoid hardcoding URLs
Automatically updates if the route changes
Makes your code cleaner, safer, and easier to maintain

#Q17 How does Flask handle static files (CSS, JavaScript, etc.)?
In Flask, static files like CSS, JavaScript, and images are served from a special folder named static/. Flask knows to look there automatically.
Put CSS/JS/images in /static/	        Flask can serve them
Use url_for('static', filename='...')	Dynamically links them in HTML
No extra route needed	                Flask handles it automatically

#Q18 What is an API specification, and how does it help in building a Flask API?
An API specification is a detailed blueprint that defines how an API should behave — including what endpoints exist, how to call them, 
what data to send, and what to expect in return.
How does help in building a flask API:
Guides development	        Acts like a map for writing routes and logic in Flask
Easier testing	            Makes it clear what inputs and outputs to expect
Team collaboration	        Backend and frontend developers can work independently
Auto-generated docs	        Tools like Swagger UI or Postman can generate documentation
Validation & standards	    Ensures the API is consistent 

#Q19 What are HTTP status codes, and why are they important in a Flask API?
HTTP status codes are three-digit numbers that a web server (like Flask) sends back to a client (browser, app, or API user) to indicate the
result of a request.
Why Are They Important:
Tell the client what happened (success, error, etc.)
Help with debugging and testing
Improve API reliability and communication
Guide front-end behavior (e.g., show success or error messages)

#Q20 How do you handle POST requests in Flask?
In Flask, you handle a POST request by:
    
1.Defining a route with the POST method.
2.Using Flask’s request object to access the data sent by the client.
    
#Q21 How would you secure a Flask API?
Securing a Flask API is essential to protect your data, prevent unauthorized access, and avoid malicious attacks.
1.Auth & tokens	            Allow only trusted users
2.HTTPS	                    Encrypt data
3.Input validation	        Block bad or malicious data
4.Rate limiting          	Stop spamming or abuse
5.CORS	Limit               who can use your API
6.Error handling	        Don’t leak info
7.Secure configs/secrets	Don’t hardcode sensitive data

#Q22 What is the significance of the Flask-RESTful extension?
Flask-RESTful is a popular Flask extension that simplifies the creation of RESTful APIs in Flask by providing tools and structure for 
building web services more easily and cleanly.

#Q23 What is the role of Flask’s session object?
The session object in Flask is used to store data across multiple requests made by the same user (client). It lets you keep track of user-specific 
information between pages — like login status, preferences, or temporary data.



```
