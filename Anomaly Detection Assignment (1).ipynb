```python
#Unsupervised Learning, Anomaly Detection, and Temporal Analysis

#Question 1 : What is Dimensionality Reduction? Why is it important in machine Learning ?

Dimensionality Reduction means reducing the number of input features (variables) in a dataset while keeping as much useful information as possible.

In simple terms ‚Äî

It‚Äôs like summarizing a long book into a few pages without losing the main story.

For example:
If your dataset has 100 features (columns), but only 10 of them are really useful for prediction, you can use dimensionality reduction to keep those 10 and remove the rest.

 Why is it important in Machine Learning?

Improves Model Performance
Too many features can make the model confused or slow (this is called the curse of dimensionality).
Reducing dimensions helps the model learn faster and more accurately.

Reduces Overfitting
Fewer irrelevant features ‚Üí model focuses on important patterns ‚Üí better generalization to new data.

Speeds Up Computation
Less data ‚Üí faster training and prediction.

Makes Data Easier to Visualize
You can visualize high-dimensional data by reducing it to 2D or 3D using methods like PCA (Principal Component Analysis).

 Common Techniques for Dimensionality Reduction
Method	Description
PCA (Principal Component Analysis)	Converts many features into fewer ‚Äúprincipal components‚Äù that explain most of the variance.
LDA (Linear Discriminant Analysis)	Reduces dimensions while keeping class information (used for classification).
t-SNE / UMAP	Used mainly for visualization of high-dimensional data.
```

```python
#Question 2: Name and briefly describe three common dimensionality reduction techniques..

Three Common Dimensionality Reduction Techniques
1. Principal Component Analysis (PCA)

What it does:
PCA transforms the original features into a new set of features called principal components.
These components capture the maximum variance (most important information) in the data.

Why it‚Äôs used:
It helps reduce the number of features while keeping most of the useful information.

Example use:
Used in image compression and speeding up machine learning models.

2. Linear Discriminant Analysis (LDA)

What it does:
LDA reduces dimensions while preserving the class (label) information.
It finds feature combinations that best separate different classes.

Why it‚Äôs used:
Mainly for classification problems (e.g., face recognition).

Example use:
Used when you want to reduce data but still keep class differences clear.

3. t-SNE (t-Distributed Stochastic Neighbor Embedding)

What it does:
t-SNE is a non-linear technique that reduces data into 2D or 3D for visualization.
It focuses on keeping similar data points close together.

Why it‚Äôs used:
It‚Äôs great for visualizing high-dimensional data, like word embeddings or image features.

Example use:
Used to visualize clusters of data points (e.g., different types of handwritten digits).
```

```python
#Question 3: What is clustering in unsupervised learning? Mention three popular clustering algorithms


Clustering is a technique in unsupervised learning where we group similar data points together based on their features ‚Äî without using any labels.

In simple words:

Clustering means organizing data into groups (clusters) so that data points in the same group are similar, and those in different groups are different.

Example:

Imagine you have data about customers ‚Äî their age, income, and spending habits.
You don‚Äôt know their group labels, but clustering can automatically group them into:

Group 1: High income, high spending

Group 2: Low income, low spending

Group 3: Young customers, moderate spending

Each group is a cluster.

üîç Three Popular Clustering Algorithms
Algorithm	Description	Best Used For
1. K-Means Clustering	Divides data into K clusters by minimizing the distance between data points and their cluster centers (centroids).	
    When clusters are spherical and evenly sized.
2. Hierarchical Clustering	Builds a tree-like structure (called a dendrogram) of clusters by continuously merging or splitting them.
    When you want to understand the hierarchical relationships between clusters.
3. DBSCAN (Density-Based Spatial Clustering of Applications with Noise)	Groups together points that are close in density and marks outliers as noise.	
    When clusters are irregularly shaped or when there is noise in the data.
```

```python
# Question 4: Explain the concept of anomaly detection and its significance.

What is Anomaly Detection?

Anomaly Detection (also called Outlier Detection) is a process of finding data points that are very different or unusual compared to the rest of the data.

In simple words:

It helps us spot anything that doesn‚Äôt fit the normal pattern.

 Example:

Imagine you‚Äôre monitoring transactions in a bank:

Most people spend ‚Çπ1,000‚Äì‚Çπ5,000 per day.

Suddenly, one account shows a ‚Çπ1,00,000 transaction ‚Äî that‚Äôs an anomaly!

So, anomaly detection would help flag this unusual transaction for further review.

 Why is Anomaly Detection Important?

Fraud Detection
Detects unusual credit card or banking activities.

Network Security
Identifies suspicious behavior in computer networks (like hacking or malware).

Manufacturing
Detects faulty machines or defective products early.

Healthcare
Finds abnormal patient readings (e.g., sudden heart rate spikes).

Data Cleaning
Removes incorrect or noisy data before model training.

‚öôÔ∏è How It Works (Basic Idea)

Anomaly detection algorithms learn what ‚Äúnormal‚Äù data looks like and then:

Mark points that deviate a lot from this normal pattern as anomalies or outliers.

Common Techniques
Method	Description
Statistical Methods	Detects points that lie far from the average (e.g., z-score, IQR).
Machine Learning Methods	Uses algorithms like Isolation Forest, One-Class SVM, or Autoencoders.
```

```python
# Question 5: List and briefly describe three types of anomaly detection techniques.

Three Types of Anomaly Detection Techniques
1. Statistical Methods

Idea:
These methods assume that normal data follows a certain statistical pattern (like a normal distribution).
Any data point that lies far from the mean or expected range is considered an anomaly.

Examples:

Z-score method (checks how far a value is from the mean)

Interquartile Range (IQR) (detects outliers using quartiles)

Use Case:
Simple datasets with numeric values (e.g., detecting unusually high temperatures).

2. Machine Learning‚ÄìBased Methods

Idea:
These methods learn patterns in normal data and flag points that don‚Äôt fit the learned model.

Examples:

Isolation Forest: Isolates anomalies faster because they are few and different.

One-Class SVM: Learns the boundary around normal data points.

Autoencoders: Neural networks that reconstruct normal data and detect high reconstruction errors as anomalies.

Use Case:
Complex datasets (e.g., fraud detection, network security).

3. Density-Based Methods

Idea:
These methods assume that normal data points are close together (high density), while anomalies are far away (low density).

Examples:

DBSCAN (Density-Based Spatial Clustering of Applications with Noise)

LOF (Local Outlier Factor) ‚Äì measures how isolated a data point is.

Use Case:
Data with irregular shapes or clusters, where distance-based assumptions don‚Äôt work well.
```

```python
# Question 6: What is time series analysis? Mention two key components of time series data?

Time Series Analysis is the process of studying data collected over time to identify meaningful patterns, trends, and relationships, and to make future predictions.

In simple terms:

It‚Äôs about analyzing how something changes with time ‚Äî like stock prices, temperature, sales, or website traffic.

Example:

Daily stock prices of a company

Monthly electricity consumption

Yearly rainfall data

Each observation is time-dependent, which makes it a time series.

Two Key Components of Time Series Data
1. Trend

The overall direction in which data is moving over a long period.

It can be upward, downward, or stable.

Example: A company‚Äôs sales increasing year by year.

2. Seasonality

Regular, repeating patterns that occur at fixed intervals (like daily, monthly, or yearly).

Example: Higher ice cream sales in summer, lower in winter.
```

```python
#Question 7: Describe the difference between seasonality and cyclic behavior in time series

Difference Between Seasonality and Cyclic Behavior in Time Series

Both seasonality and cyclic behavior are patterns seen in time series data ‚Äî
but they differ in how regularly they occur and what causes them.

Feature	Seasonality	Cyclic Behavior
Definition	Repeating patterns that occur at fixed and regular time intervals.	Patterns that occur in irregular and long-term cycles.
Time Period	Happens at specific intervals ‚Äî e.g., daily, monthly, or yearly.	Has no fixed duration; cycles can vary in length.
Cause	Caused by seasonal factors like weather, holidays, or time of year.	Caused by economic or business conditions, not by the calendar.
Example	Ice cream sales rise every summer and fall every winter.	The business cycle ‚Äî periods of economic growth followed by slowdown.
Predictability	Predictable ‚Äî happens regularly.	Less predictable ‚Äî depends on external factors.
 In Simple Terms:

Seasonality = Regular, short-term, calendar-based pattern
(e.g., every December, every Monday, every summer)

Cyclic Behavior = Irregular, long-term economic or natural cycle
(e.g., business boom and recession)
```

```python
# Question 8: Write Python code to perform K-means clustering on a sample dataset.(Include your Python code and output in the code box below.

# K-Means Clustering Example

# Step 1: Import required libraries
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Step 2: Create a sample dataset
X, y = make_blobs(
    n_samples=200,      # number of data points
    centers=3,          # number of clusters
    random_state=42,    # ensures same result every time
    cluster_std=1.0     # spread of clusters
)

# Step 3: Apply K-Means Clustering
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)

# Step 4: Get the cluster labels and centroids
labels = kmeans.labels_
centroids = kmeans.cluster_centers_

# Step 5: Visualize the clusters
plt.figure(figsize=(8, 5))
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=50)
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', s=200, marker='X')
plt.title("K-Means Clustering Example")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()

# Step 6: Display cluster centers
print("Cluster Centers:\n", centroids)
```

```python
#Question 9: What is inheritance in OOP? Provide a simple example in Python.

What is Inheritance in OOP?

Inheritance is one of the main concepts in Object-Oriented Programming (OOP).
It allows a child class (also called a subclass) to reuse the properties and methods of a parent class (also called a superclass).

In simple words:

Inheritance means one class can use or extend the behavior of another class.

üí° Why Use Inheritance?

To reuse code instead of writing it again.
 To make the program organized and easier to maintain.
To add or modify features of an existing class.
    
```

```python
# Parent class
class Animal:
    def speak(self):
        print("Animals can make sounds")

# Child class
class Dog(Animal):
    def bark(self):
        print("Dog barks: Woof! Woof!")

# Create an object of the child class
my_dog = Dog()

# Access methods from both classes
my_dog.speak()  # Inherited from Animal
my_dog.bark()   # Defined in Dog
```

```python
#Question 10: How can time series analysis be used for anomaly detection?

How Time Series Analysis Can Be Used for Anomaly Detection

Time Series Anomaly Detection means finding unusual or unexpected patterns in data that change over time.

In simple words:

It helps identify when something happens that doesn‚Äôt follow the usual time pattern ‚Äî like a sudden drop or spike.

üí° Example:

A website‚Äôs traffic suddenly drops to half ‚Äî might mean a system issue.

A sensor suddenly reports a very high temperature ‚Äî could mean equipment failure.

A bank transaction happens at an unusual time or amount ‚Äî possible fraud.

‚öôÔ∏è How It Works (Steps):

Collect Time Series Data
(e.g., sales per day, temperature per hour, or network usage per minute)

Model Normal Behavior
Use time series analysis methods like:

Moving Average

ARIMA (AutoRegressive Integrated Moving Average)

Exponential Smoothing
These models learn the normal trend and seasonality in the data.

Detect Deviations
Compare actual values with predicted or expected values.

If the difference (error) is too large ‚Üí it‚Äôs marked as an anomaly
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Create sample time series data
np.random.seed(42)
time = pd.date_range(start='2023-01-01', periods=50)
values = np.random.normal(100, 5, size=50)
values[25] = 140  # introduce an anomaly

# Simple threshold-based anomaly detection
threshold = 120
anomalies = values > threshold

# Plot results
plt.figure(figsize=(8,5))
plt.plot(time, values, label='Data')
plt.scatter(time[anomalies], values[anomalies], color='red', label='Anomaly', s=100)
plt.title("Time Series Anomaly Detection Example")
plt.xlabel("Date")
plt.ylabel("Value")
plt.legend()
plt.show()
```
