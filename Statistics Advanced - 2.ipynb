```python
# Statistics Advanced - 2

#Q1 What is hypothesis testing in statistics?
Hypothesis testing is a method in statistics that helps us decide if what we observe in our data is real or just due to random chance.
Think of it as a court trial:
Null Hypothesis (H‚ÇÄ) = "The person is innocent" (no effect, no difference).
Alternative Hypothesis (H‚ÇÅ or Ha) = "The person is guilty" (there is an effect, there is a difference).
The evidence = your sample data.
The judge/jury = statistical test.

#Steps of Hypothesis Testing

1.State the hypotheses

H‚ÇÄ (Null): No effect / No difference.

H‚ÇÅ (Alternative): There is an effect / There is a difference.

2.Choose a significance level (Œ±)

Commonly 0.05 (5%) ‚Üí This means you‚Äôre okay with a 5% chance of wrongly rejecting H‚ÇÄ.

3.Collect data and perform a test

Example: t-test, z-test, chi-square test, etc.

4.Calculate the p-value

p-value = probability of observing your data (or more extreme) if H‚ÇÄ were true.

5.Make a decision

If p-value ‚â§ Œ± ‚Üí Reject H‚ÇÄ (evidence supports H‚ÇÅ).

If p-value > Œ± ‚Üí Fail to reject H‚ÇÄ (not enough evidence).
```

```python
#Q2 What is the null hypothesis, and how does it differ from the alternative Hypothesis?
Null Hypothesis (H‚ÇÄ)
The starting assumption in hypothesis testing.
It usually says:
"Nothing is happening."
"No difference."
"No effect."
Think of it as the default belief.
Example:
A company says their drug has no effect on blood pressure.
H‚ÇÄ: The drug does not change blood pressure.

Alternative Hypothesis (H‚ÇÅ or Ha)
This is what you want to test/prove.
It says:
"Something is happening."
"There is a difference."
"There is an effect."
It‚Äôs the opposite of the null hypothesis.
Example:
H‚ÇÅ: The drug does change blood pressure.
```

```python
#Q3 Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.

What is Significance Level (Œ±)?

The significance level is a threshold (a cutoff) that we set before doing the test.
It tells us how much risk we are willing to take of wrongly rejecting the null hypothesis (H‚ÇÄ).
Most common choice: Œ± = 0.05 (5%)
This means:
We accept a 5% chance of saying something is happening (rejecting H‚ÇÄ) when in reality, nothing is happening (H‚ÇÄ is true).

Role of Œ± in Hypothesis Testing
When we run a test, we calculate a p-value (probability of observing the data if H‚ÇÄ is true).
If p ‚â§ Œ± ‚Üí Reject H‚ÇÄ (evidence supports H‚ÇÅ).
If p > Œ± ‚Üí Fail to reject H‚ÇÄ (not enough evidence to support H‚ÇÅ).
So, Œ± is like the cutoff line for making the decision.

üîπ Example
Suppose a new medicine claims to lower blood pressure.

H‚ÇÄ: The medicine has no effect.
H‚ÇÅ: The medicine lowers blood pressure.
Œ± = 0.05
After testing, you get p = 0.03.
Since 0.03 < 0.05, reject H‚ÇÄ ‚Üí The medicine likely works.
But if p = 0.08:
Since 0.08 > 0.05, fail to reject H‚ÇÄ ‚Üí Not enough evidence.
```

```python
#Q4. What are Type I and Type II errors? Give examples of each?

Type I Error (False Positive)
This happens when we reject the null hypothesis (H‚ÇÄ) even though it is actually true.
In other words: we see an ‚Äúeffect‚Äù that is really just due to chance.
Analogy (Courtroom):
We wrongly declare an innocent person guilty.
Example:
A company claims their drug has no effect on blood pressure.
H‚ÇÄ: The drug has no effect.
H‚ÇÅ: The drug lowers blood pressure.
If in reality the drug does nothing, but our test says it works ‚Üí Type I Error.
Probability of making a Type I error = Œ± (significance level).
If Œ± = 0.05, we‚Äôre allowing a 5% chance of this mistake.

Type II Error (False Negative)
This happens when we fail to reject the null hypothesis (H‚ÇÄ) even though it is actually false.
In other words: we miss a real effect.
Analogy (Courtroom):
We wrongly declare a guilty person innocent.
Example:
Using the same drug case:
If the drug really does lower blood pressure, but our test fails to detect it (we conclude ‚Äúno effect‚Äù) ‚Üí Type II Error.
Probability of making a Type II error = Œ≤ (beta).
And 1 - Œ≤ = power of the test (ability to detect a true effect).
```

```python
#Q5 What is the difference between a Z-test and a T-test? Explain when to use each?
Z-test
When used:
Large sample size (n > ~30).
The population standard deviation (œÉ) is known.
Data is approximately normally distributed (or sample is large enough for Central Limit Theorem).
Why: With large samples or known œÉ, the sampling distribution of the mean is well-approximated by the standard normal distribution (Z).
Example:
A factory says their bolts have a mean length of 5 cm with a known œÉ = 0.2 cm. You sample 50 bolts and want to check if the mean is really 5 cm ‚Üí 
Z-test.

T-test
When used:
Small sample size (n < ~30).
The population standard deviation (œÉ) is unknown (you estimate it using the sample standard deviation, s).
Data should be approximately normal (especially important for small samples).
Why: With small samples, there‚Äôs more uncertainty in estimating œÉ. The t-distribution (which has ‚Äúfatter tails‚Äù than normal) accounts for this extra 
variability.
Example:
A nutritionist wants to test if a new diet changes average weight. They collect data from 15 people and don‚Äôt know the population œÉ ‚Üí T-test.
```

```python
#Q7 Implement hypothesis testing using Z-statistics for a sample dataset in
#Python. Show the Python code and interpret the results.
#sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
#50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
#50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
#50.3, 50.4, 50.0, 49.7, 50.5, 49.9]


import numpy as np
from scipy.stats import norm

# Sample data
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

# Parameters
mu_0 = 50     # hypothesized population mean
alpha = 0.05  # significance level

# Sample statistics
sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data, ddof=1)  # sample standard deviation
n = len(sample_data)

# Z-statistic
z_stat = (sample_mean - mu_0) / (sample_std / np.sqrt(n))

# p-value (two-tailed test)
p_value = 2 * (1 - norm.cdf(abs(z_stat)))

# Critical Z for two-tailed test
z_critical = norm.ppf(1 - alpha/2)

# Print results
print(f"Sample Mean: {sample_mean:.4f}")
print(f"Sample Std Dev: {sample_std:.4f}")
print(f"Z-statistic: {z_stat:.4f}")
print(f"P-value: {p_value:.4f}")
print(f"Z-critical (two-tailed, alpha=0.05): ¬±{z_critical:.4f}")

# Decision
if abs(z_stat) > z_critical:
    print("Reject H‚ÇÄ: The sample mean is significantly different from 50.")
else:
    print("Fail to reject H‚ÇÄ: No significant evidence that the mean differs from 50.")
```

```python
#Q8 Write a Python script to simulate data from a normal distribution and
#calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# 1. Simulate normal distribution data
np.random.seed(42)  # for reproducibility
mu, sigma = 50, 5   # true mean and std dev
n = 100             # sample size
data = np.random.normal(mu, sigma, n)

# 2. Calculate sample statistics
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)

# 95% Confidence Interval for the mean
alpha = 0.05
z_critical = norm.ppf(1 - alpha/2)  # ~1.96
margin_of_error = z_critical * (sample_std / np.sqrt(n))
ci_lower = sample_mean - margin_of_error
ci_upper = sample_mean + margin_of_error

# Print results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")

# 3. Plot the data
plt.figure(figsize=(8,5))
plt.hist(data, bins=15, edgecolor='black', alpha=0.7)
plt.axvline(sample_mean, color='red', linestyle='dashed', linewidth=2, label=f"Mean = {sample_mean:.2f}")
plt.axvline(ci_lower, color='green', linestyle='dashed', linewidth=2, label=f"95% CI Lower = {ci_lower:.2f}")
plt.axvline(ci_upper, color='green', linestyle='dashed', linewidth=2, label=f"95% CI Upper = {ci_upper:.2f}")
plt.title("Histogram of Simulated Normal Data")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

```python
#Q9 Question 9: Write a Python function to calculate the Z-scores from a dataset and
#visualize the standardized data using a histogram. Explain what the Z-scores representin terms of standard deviations from the mean.

import numpy as np
import matplotlib.pyplot as plt

def calculate_z_scores(data):
    """
    Calculate Z-scores for a dataset and visualize with histogram.
    """
    # Convert to numpy array
    data = np.array(data)
    
    # Calculate mean and std deviation
    mean = np.mean(data)
    std = np.std(data, ddof=1)  # sample std
    
    # Compute Z-scores
    z_scores = (data - mean) / std
    
    # Print first few Z-scores
    print("First 5 Z-scores:", z_scores[:5])
    
    # Plot histogram
    plt.figure(figsize=(8,5))
    plt.hist(z_scores, bins=10, edgecolor="black", alpha=0.7)
    plt.axvline(0, color='red', linestyle='dashed', linewidth=2, label="Mean (Z=0)")
    plt.title("Histogram of Standardized Data (Z-scores)")
    plt.xlabel("Z-score (Standard Deviations from Mean)")
    plt.ylabel("Frequency")
    plt.legend()
    plt.show()
    
    return z_scores

# Example dataset
data = [10, 12, 13, 15, 18, 19, 21, 22, 23, 25]
z_scores = calculate_z_scores(data)
```
